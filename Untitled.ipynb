```python
# ================================================================
# Monte Carlo Simulation for Option Pricing (using real data)
# Author: Naman Rajput
# ================================================================

import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.optimize import brentq
from tqdm import tqdm

plt.style.use("seaborn-v0_8")


# Downloading data

ticker = "AAPL"     # change to SPY, TSLA, etc.
start_date = "2024-01-01"
end_date   = "2025-10-24"

data = yf.download(ticker, start="2023-01-01", end="2025-06-01")
data.to_csv("AAPL_prices.csv")
data.tail()
```

```python
#  Compute historical volatility

if isinstance(data.columns, pd.MultiIndex):
    data.columns = data.columns.get_level_values(0)

# Compute log returns
log_returns = np.log(data["Close"] / data["Close"].shift(1)).dropna()

# Convert to scalar values
sigma_daily = float(log_returns.std(ddof=1))
sigma_annual = float(sigma_daily * np.sqrt(252))
S0 = float(data["Close"].iloc[-1])

print(f"Current Price S0 = {S0:.2f}")
print(f"Annualized Volatility (hist) = {sigma_annual:.4f}")
```

```python
#  Get option chain (current snapshot)

t = yf.Ticker(ticker)
expiries = t.options
print("Available expiries:", expiries[:5])

# choose nearest expiry
expiry = expiries[0]
chain = t.option_chain(expiry)
calls = chain.calls
puts = chain.puts
calls.head()
```

```python
#  Preprocess and select ATM strikes

calls["mid"] = (calls["bid"] + calls["ask"]) / 2
calls = calls.dropna(subset=["mid"])
atm_strike = min(calls["strike"], key=lambda x: abs(x - S0))

K_values = sorted(calls["strike"].unique())[:8]
print("Sample strikes:", K_values)
```

```python
#  Helper: Black-Scholes and implied vol solver

def bs_call_price(S, K, r, sigma, T, q=0.0):
    if sigma <= 0 or T <= 0:
        return max(0.0, S - K*np.exp(-r*T))
    d1 = (np.log(S/K) + (r - q + 0.5*sigma**2)*T) / (sigma*np.sqrt(T))
    d2 = d1 - sigma*np.sqrt(T)
    return S*np.exp(-q*T)*norm.cdf(d1) - K*np.exp(-r*T)*norm.cdf(d2)

def implied_vol_call(S, K, r, T, q, market_price):
    f = lambda sigma: bs_call_price(S,K,r,sigma,T,q) - market_price
    try:
        return brentq(f, 1e-6, 5.0)
    except ValueError:
        return np.nan
```

```python
#  Estimate implied vols for strikes

r = 0.04       # risk-free rate
q = 0.0
T_days = (pd.to_datetime(expiry) - data.index[-1]).days
T = T_days / 252

ivs = []
for k in tqdm(K_values):
    mp = calls.loc[calls["strike"]==k, "mid"].mean()
    iv = implied_vol_call(S0,k,r,T,q,mp)
    ivs.append(iv)

vol_df = pd.DataFrame({"Strike":K_values,"ImpliedVol":ivs})
vol_df
```

```python
#  Monte Carlo pricer (antithetic + control variate)

def mc_call_price(S0, K, r, q, sigma, T, n_paths=100_000, n_steps=50):
    dt = T/n_steps
    half = n_paths//2
    Z = np.random.normal(size=(half, n_steps))
    Z = np.vstack([Z,-Z])                          # antithetic
    increments = (r-q-0.5*sigma**2)*dt + sigma*np.sqrt(dt)*Z
    logS = np.log(S0) + increments.cumsum(axis=1)
    S_T = np.exp(logS[:,-1])
    payoff = np.maximum(S_T - K, 0.0)

    # control variate using BS closed-form
    bs = bs_call_price(S0,K,r,sigma,T,q)
    control = S_T                               # control variable with known E[S_T]
    EY = S0*np.exp((r-q)*T)
    cov = np.cov(payoff,control,ddof=1)[0,1]
    b_opt = cov / np.var(control,ddof=1)
    payoff_cv = payoff - b_opt*(control - EY)

    disc = np.exp(-r*T)
    price = disc * payoff_cv.mean()
    stderr = disc * payoff_cv.std(ddof=1) / np.sqrt(n_paths)
    ci95 = (price - 1.96*stderr, price + 1.96*stderr)
    return price, stderr, ci95
```

```python
#  Compute MC and BS prices across strikes

res = []
for i,(K,sig) in enumerate(zip(K_values,ivs)):
    if np.isnan(sig):
        sig = sigma_annual
    mc_p, mc_err, ci = mc_call_price(S0,K,r,q,sig,T)
    bs_p = bs_call_price(S0,K,r,sig,T,q)
    market = calls.loc[calls["strike"]==K,"mid"].mean()
    res.append([K,sig,bs_p,mc_p,mc_err,market,ci[0],ci[1]])

result = pd.DataFrame(res,columns=["Strike","Sigma","BS_Price","MC_Price",
                                   "MC_Error","Market","CI_low","CI_high"])
result
```

```python
#  Visualization
plt.figure(figsize=(10,6))
plt.plot(result["Strike"],result["BS_Price"],"o-",label="Black-Scholes")
plt.plot(result["Strike"],result["MC_Price"],"s-",label="Monte Carlo")
plt.scatter(result["Strike"],result["Market"],color="red",label="Market mid")
plt.xlabel("Strike"); plt.ylabel("Option Price"); plt.title(f"{ticker} Call Prices (expiry {expiry})")
plt.legend(); plt.grid(True); plt.show()
```

```python
#  Error metrics

result["AbsError"] = (result["MC_Price"] - result["Market"]).abs()
result["RelError_%"] = 100*result["AbsError"]/result["Market"]
print("Mean Abs Error:", result["AbsError"].mean())
print("Mean Rel Error %:", result["RelError_%"].mean())
result[["Strike","Market","BS_Price","MC_Price","RelError_%"]]
```

```python
# Save results

result.to_csv("results_option_pricing.csv",index=False)
print("Results saved to results_option_pricing.csv")
```
